```python
import pandas as pd
from lxml import etree
from bs4 import BeautifulSoup
import requests
import time
snake_df = pd.read_csv("WHC_html.csv")
```

```python
snake_df.info()
```

```python
## use for filtering below Extracting data
def fun(x):
    return x[1]
```

```python
big_dict = {}  # it contains all snakes species name and their value sequentially
full_dict = {}  # it contains snake species name as well as venom type and and their value
unidentify = []
for i in range(0, 2819):

    try:
        x = BeautifulSoup(snake_df['content_html'][i], 'html.parser')

        table_venom = x.find_all("table")[13]

        sp = x.find_all("td", attrs={"class": "page_title_brown"})[0].text  # sp is snake species name

        mini_dict = {}
        for j in table_venom.find_all('b'):  # j:venom type in html response

            # next_sibling is value for j
            m = j.findNextSibling("br").next_sibling
            mini_dict[j.text] = m

        # some of the response doesn't contain Average Venom Qty so we need to assign None value for completing
        # our dictionary
        if "Average Venom Qty" not in mini_dict.keys():
            mini_dict["Average Venom Qty"] = None

        # sorting venom value by their type because in building pandas dataframe we need to follow sequence
        u = map(fun, sorted(mini_dict.items(), key=lambda x: (x[0])))
        big_dict[sp] = list(u)

        full_dict[sp] = mini_dict
    except:
        # something bad response filter out and their number
        unidentify.append(i)
```

```python
## just save dictionary for not running above process again and again 
try:
    geeky_file = open('full_dict.txt', 'wt')
    geeky_file.write(str(full_dict))
    geeky_file.close()

except:
    print("Unable to write to file")
```

```python
## making function for further using 
def unique_word_list_values(df_data):
    unique_columns_values = {} ## we needs to make dictionary with venom type and all unique value for this venom type
    unique_word_list = {} ## all unique word and it's frequency in whole dataframe
    for i in df_data.columns:
        if i not in ["snake name", "Average Venom Qty"]:
            temp=list(df_data[i].unique())

            temp1=list(df_data[i])
            unique_columns_values[i] = temp
            for i in temp1:
                unique_word_list[i]=unique_word_list.get(i,0)+1
    return [unique_columns_values,unique_word_list]
            
```

```python
## i think same non-venoms word have too many index so i build this function 
def return_Non_venom_undecided_frame(df_data,word):
    original_df=df_data.copy(deep=True)
    df_data=df_data.drop(labels=["snake name","class"],axis=1)
    Non_venom_index=list(df_data[(df_data==word).all(1)].index)
    undecided_index=list(df_data[~(df_data==word).all(1)].index)
    Non_venom_csv=original_df.loc[df_data.index.isin(Non_venom_index)]
    undecided_label=original_df.loc[~df_data.index.isin(Non_venom_index)]
    return [Non_venom_csv,undecided_label]    
    
```
