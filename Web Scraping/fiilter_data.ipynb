```python
big_dict = {}
full_dict = {}
unidentify = []
for i in range(0, 2819):

    try:
        x = BeautifulSoup(snake_df['content_html'][i], 'html.parser')

        table_venom = x.find_all("table")[13]

        sp = x.find_all("td", attrs={"class": "page_title_brown"})[0].text

        keys = []
        mini_dict = {}
        for j in table_venom.find_all('b'):
            #         keys.append(i.text)
            m = j.findNextSibling("br").next_sibling
            mini_dict[j.text] = m

    #         print(j.text,j.findNextSibling("br").next_sibling)

        if "Average Venom Qty" not in mini_dict.keys():
            mini_dict["Average Venom Qty"] = None

        u = map(fun, sorted(mini_dict.items(), key=lambda x: (x[0])))
        big_dict[sp] = list(u)
        full_dict[sp] = mini_dict
    except:
        unidentify.append(i)
```
