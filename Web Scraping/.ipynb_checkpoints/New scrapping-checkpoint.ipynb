## Round 1

```python
list_1_to_500 = []
for i in range(1, 5):
    id_snake = str(i)
    length=len(id_snake)
    if length <=4:
        x = (4-length)*"0"+id_snake
        print()

        page = requests.get(
            "http://www.toxinology.com/fusebox.cfm?fuseaction=main.snakes.display&id=SN"+x)
        list_1_to_500.append(page)
        time.sleep(2)
print("done")
```

```python
BeautifulSoup(list_1_to_500[0].content)
```

```python
data_1_to_500=pd.DataFrame(data=np.array(list_1_to_500),columns=["0_to_500_response"])
```

```python
for i,j in enumerate(list_1_to_500):
     data_1_to_500.loc[i,"content"]=j.content
```

## round 2

```python
list_1000_to_1999 = []
for i in range(1000, 1999):
    id_snake = str(i)
    length=len(id_snake)
    if length <=4:
        x = (4-length)*"0"+id_snake
        page = requests.get(
            "http://www.toxinology.com/fusebox.cfm?fuseaction=main.snakes.display&id=SN"+x)
        list_1000_to_1999.append(page)
        time.sleep(2)
    print(i)
print("done")
```

```python
data_1000_to_1999=pd.DataFrame(data=np.array(list_1000_to_1999),columns=["1000_to_response"])
```

```python
for i, j in enumerate(list_1000_to_1999):
#     data_1000_to_1999.loc[i, "0_to_500_response"] = j
    data_1000_to_1999.loc[i, "content"] = j.content
   
```

```python
data_1000_to_1999.rename(columns={"1000_to_response":"1000_to_1998"})
```

## Round 4 merge all data

```python
df1=pd.read_csv("WHC_Data.csv")
df2=pd.read_csv("WHC2.csv")
df3=pd.read_csv("WHC3.csv")
```
