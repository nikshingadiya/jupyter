```python
import pandas as pd 
```

## venom two file combine 

```python
venom_0=pd.read_csv("./Venom/venom_data_frame_0.csv")
venom_1=pd.read_csv("./Venom/venom_data_frame_1.csv")
```

```python
venom_0.head()
```

```python
venom_1.head()
```

```python
venom=pd.concat([venom_0,venom_1])
```

```python
venom=venom.reset_index(0,drop=True)
```

```python
venom.info()
```

```python
venom["snake_family"]=venom["snake name"].apply(lambda x: x.split(" ")[0])
```

```python
sorted(venom['snake name'].unique())
```

## Non venom combine 

```python
non_venom_1=pd.read_csv("./Nonvenom/Non_venom_csv_1.csv")
non_venom_2=pd.read_csv("./Nonvenom/Non_venom_csv_2.csv")
```

```python
non_venom=pd.concat([non_venom_1,non_venom_2])
```

```python
non_venom
```

```python
non_venom=non_venom.reset_index(0,drop=True)
```

```python
non_venom
```

```python
non_venom["snake_family"]=non_venom["snake name"].apply(lambda x: x.split(" ")[0])
```

```python
non_venom['class']=non_venom["class"].replace({"Not":0})
```

```python
## Final csv build 
```

```python
snake_classified_final_venom_and_non_venom=pd.concat([venom,non_venom])
```

```python
snake_classified_final_venom_and_non_venom.head()
```

```python
snake_classified_final_venom_and_non_venom.info()
```

```python
snake_classified_final_venom_and_non_venom.insert(1,"higher_snake_family",snake_classified_final_venom_and_non_venom["snake_family"])
```

```python
snake_classified_final_venom_and_non_venom
```

```python
snake_classified_final_venom_and_non_venom=snake_classified_final_venom_and_non_venom.rename(columns={"snake name":"snake_sub_family"})
```

```python
snake_classified_final_venom_and_non_venom=snake_classified_final_venom_and_non_venom.reset_index(0,drop=True)
```

```python
snake_classified_final_venom_and_non_venom.head(100)
```

```python
import os
```

```python
os.mkdir("./final_snake_poisonous_non_poisonous_dir")
```

```python
snake_classified_final_venom_and_non_venom.to_csv("./final_snake_poisonous_non_poisonous_dir/snake_classified_final_venom_and_non_venom.csv",index=False)
```

```python
snake_label_only=snake_classified_final_venom_and_non_venom[["snake_sub_family","higher_snake_family","class"]]
```

```python
snake_label_only=snake_label_only.reset_index(0,drop=True)
```
