```python
i = 0
temp=i
only_key_and_number={}

# key_diffrence_value={}
# list_diffrence_frame=[]
while(frames.isOpened()):
    ret, frame = frames.read()
    if ret==False: 
        break
    
    if i%49==0:
        temp=i
        print(i)
        only_key_and_number[i]=frame      
        cv2.imwrite(f"./50th_image_key/key_{i}.jpg",frame)  ## key frame image store 
        
        os.mkdir(f'./key_diffrence/key_{i}') ## key frame folder 
    else:
        print(temp,i)
        diffrence=substraction_frame(frame,only_key_and_number[temp])
#         list_diffrence_frame.append((temp,diffrence))
        cv2.imwrite(f'./key_diffrence/key_{temp}/{temp}_{i}.jpg',diffrence) ## diffrence image store 
    

    i+=1
    
    
    

# frames.release() ## destroy frames 
# cv2.destroyAllWindows()
```

## Merging Part

```python
key_image_list=os.listdir("./50th_image_key/") ## get key image list 
```

```python
key_base_path="./50th_image_key/"  ## key_base_path  for further using 
diffrence_base_path="./key_diffrence/" ## diffrence_bash_path is root folder of all  diffrence frames 
```

```python
def sorting_list(list_1):
    '''This function used for sorting file name because we needs to maintain our frame sequences '''
    key_sort={}
    for i in list_1:
         key_sort[i]=int(i.split(".")[0].split("_")[1])
    key_sort=dict(sorted(key_sort.items(),key=lambda x:x[1]))
    return list(key_sort.keys())
```

```python
key_image_list=sorting_list(key_image_list) ## key_image file name sort
```

```python
import shutil
# shutil.rmtree("./merge_frame/") ## use this line code if we have old folder 
os.mkdir("./merge_frame") ## in this folder our sequential frame arrange 
```

```python
merge_base_path="./merge_frame"  ## merge path 
```

```python
def addition_frame(diff,fk):
    return cv2.add(diff,fk)
```

```python
x = 0 ## count our image 
for i in key_image_list:
     ## printing frame sequence 
    
    img_key = cv2.imread(key_base_path+i) ## key image get 
    cv2.imwrite(f"{merge_base_path}/{x}.jpg", img_key) ## write image in merge folder 
    
    x += 1 
    print(f"{x}.jpg")
    key_folder_path = diffrence_base_path+i.split(".")[0] ## diffrence_bash_path+ key folder (Like key_0 folder)
    list_diffrence_image = sorting_list(os.listdir(key_folder_path)) ## sorting filenames 

    for j in list_diffrence_image:
        diff_img = cv2.imread(key_folder_path+"/"+j) ## get diffrence image from key_folder 

        orignal_img = addition_frame(diff_img, img_key) ## addition key+diffrence 
        cv2.imwrite(f"{merge_base_path}/{x}.jpg", orignal_img) ## write orinal_image to merge_frame
        x += 1
        print(f"{x}.jpg")
```

##  Frames to Video 

```python
frames_name=os.listdir("./merge_frame/") ## frames list get from merge_frame
```

```python
frames.get(cv2.CAP_PROP_FOURCC)
```

<b><big>frames:</big></b> orignal video frames object contain we needs orignal video property 

```python
height=int(frames.get(cv2.CAP_PROP_FRAME_HEIGHT)) ## 
width=int(frames.get(cv2.CAP_PROP_FRAME_WIDTH))
fps=frames.get(cv2.CAP_PROP_FPS)
fourcc=int(frames.get(cv2.CAP_PROP_FOURCC))
```

```python
def sorting_frames_number(frames):
    '''sorting frame merger folder files name '''
    sort_frame={}
    for i in frames:
         sort_frame[i]=int(i.split(".")[0])
    sort_frame=dict(sorted(sort_frame.items(),key=lambda x:x[1]))
    return list(sort_frame.keys())
```

```python
frames_name=sorting_frames_number(frames_name) ## sorting frames_name 
```

```python
def video_return(frames_folder_path, height, width, fps, name,fourcc):
    '''this function convert frames to video'''
    name = name + '.mp4'
#     fourcc = cv2.VideoWriter_fourcc(*'MP4V')
    out = cv2.VideoWriter(filename=name, fourcc=fourcc,
                          fps=fps, frameSize=(width, height))
    frames_name = sorting_frames_number(os.listdir(merge_base_path))
    x = 0
    ''' Use below comment code for generating short video just for checking our video build in right way '''
    
#     for i in frames_name:
#         out.write(cv2.imread(frames_folder_path+"/"+i))
#         if x == 1000:
#             break
#         x+=1

    for i in frames_name:
        out.write(cv2.imread(frames_folder_path+"/"+i))
        
    out.release() ## Note this video is bigger than the orinal_video
```

```python
video_return(merge_base_path,height,width,fps,"orinal_video",fourcc)
```
