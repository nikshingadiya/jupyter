```python
from numpy import expand_dims
from keras.applications.vgg16 import preprocess_input
import os 
import shutil
```

```python
from keras.applications.vgg16 import VGG16
# load the model
import matplotlib.pyplot as plt
from keras.models import Model
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array

model = VGG16()
# summarize the model
model.summary()
```

```python
for layer in model.layers:
	# check for convolutional layer
	if 'conv' not in layer.name:
		continue
	# get filter weights
	filters, biases = layer.get_weights()
	print(layer.name, filters.shape)
```

```python
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
f = plt.figure(figsize=(16,16))
# plot first few filters
n_filters, ix = 6, 1
for i in range(n_filters):
	# get the filter
	f = filters[:, :, :, i]
	# plot each channel separately
	for j in range(3):
		# specify subplot and turn of axis
		ax = plt.subplot(n_filters, 3, ix)
		ax.set_xticks([])
		ax.set_yticks([])
		# plot filter channel in grayscale
		plt.imshow(f[:, :, j], cmap='gray')
		ix += 1
# show the figure
plt.show()
```

```python
def model_get(i=1,model=None):
  model_1 = Model(inputs=model.inputs, outputs=model.layers[i].output)
  model_1.summary()
  return model_1
```

```python
def preprocess_image():
  img_path="/content/drive/MyDrive/Colab"+" Notebooks/VGG16/ships_sea_light_rain_69192_2560x1080.jpg"
  img = load_img(img_path, target_size=(224, 224))
  # convert the image to an array
  img = img_to_array(img)
  img = expand_dims(img, axis=0)
  # prepare the image (e.g. scale pixel values for the vgg)
  img = preprocess_input(img)
  return img
```

```python
def feaure_maps(md):
  feature_maps=md.predict(preprocess_image())
  return feature_maps
```

```python
def layer_output(features,path):
  square = 8
  ix = 1
  plt.figure(figsize=(12,12))
  for _ in range(square):
      for _ in range(square):
        ax = plt.subplot(square, square, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        # plot filter channel in grayscale
       
        
        plt.imshow(features[0, :, :, ix-1], cmap='inferno')
        plt.savefig(path+"/"+str(ix))
        # print(features[0, :, :, ix-1].shape)
        ix += 1
  # show the figure
```

```python
def model_output(base_path):
  for i in enumerate(model.layers):
      try:
        if i[0]!=0:
          model_local=model_get(i[0],model)
          
          feautres=feaure_maps(model_local)
          
          layer_path=base_path+"/"+i[1].name
          print("layer_name:=",i[1].name)
          if os.path.exists(layer_path):
              shutil.rmtree(layer_path)
          os.mkdir(layer_path)
          print("nik")
         
          layer_output(feautres,layer_path)
          

         
        else:
             pass
      except Exception  as e:
          print(e)
          break
  
        
      
    
```
