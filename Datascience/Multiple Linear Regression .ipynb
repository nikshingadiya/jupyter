```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Dataset//1.02. Multiple linear regression.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
x = data[['SAT','Rand 1,2,3']]
y = data['GPA']
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(x,y)
```

### Formula for Adjusted R-squared

$R^2_{adj} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$ 

```python
x.shape
```

```python
r2 = reg.score(x,y)

n = x.shape[0]
p = x.shape[1]

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

# Feature Selection

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1] 
p_values
```

```python
p_values.round(3)
# SAT,Rand 1,2,3
#Rand 1,2,3 is useless as >0.05
```
